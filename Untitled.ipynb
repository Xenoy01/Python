```python
s=input("Enter the string:")
s=s.lower()
af={}
for c in  s:
    if (c.isalpha()):
        af[c]=af.get(c,0)+1
plt.plot(af.keys(),af.values())
plt.xlabel("Alphabets")
plt.ylabel("Frequencies")
plt.title("Frequency of Alphabet")
plt.show()
```
